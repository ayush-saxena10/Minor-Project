### Data Preprocessing

```python
sns.heatmap(df.isnull())
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
# df_train_sample.to_excel("train_sample.xlsx")
# df_test_sample.to_excel("test_sample.xlsx")
```

```python
sc_X = sklearn.preprocessing.StandardScaler()
sc_y = sklearn.preprocessing.StandardScaler()

X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)
# y_train = sc_y.fit_transform(y_train)
# y_test = sc.transform(y_test)
```

### ANN Model Building

```python
ann = tf.keras.models.Sequential()
```

```python
# Input and first hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
# Second hidden layer
ann.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
# Output layer
ann.add(tf.keras.layers.Dense(units=1))
```

```python
# Compiling the ANN
ann.compile(optimizer='adam', loss='mean_squared_error', metrics=['mae'])
```

### Training the model

```python
history = ann.fit(X_train, y_train, batch_size=10, epochs=300)
```

```python
y_pred = ann.predict(X_test)
np.set_printoptions(precision=2)
```

```python
plt.plot(history.history['loss'])
plt.title('Mean Squared Error')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper right')
plt.show
```

```python
plt.plot(history.history['mae'])
plt.title('Mean Absolute Error')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper right')
plt.show
```
